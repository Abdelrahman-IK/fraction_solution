```python
import glob
```

```python
df = pd.DataFrame()
for file in glob.glob("data/events/*.txt")[:3]:
    with open(file) as f:
        data = f.read()
    print(data)
```

```python
df = pd.DataFrame(data.split('\n'),columns=['logs'])
```

```python
df['timestamp'] = df['logs'].apply(lambda x: x.split(' GMT')[0])
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
import ast,re
df['data'] = df['logs'].apply(lambda x: ast.literal_eval(re.search('({.+})', x).group(0)))
```

```python
df = pd.concat([df,df['data'].apply(pd.Series)], axis=1)
del df['logs']
del df['data']
```
